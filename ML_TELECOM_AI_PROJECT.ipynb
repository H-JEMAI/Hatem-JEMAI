```python
#Filter High Value Customers
#Define high-value customers as follows:
#Those who have recharged with an amount more than or equal to X, where X is the 70th percentile of the average recharge amount in the first two months (the good phase).
#There are lot of missing values for the Data and the Amt_Data Columns indicating that no recharge was done on that month
#The NaN values should be replaced by 0
df.total_rech_data_6.fillna(value=0, inplace=True)
df.total_rech_data_7.fillna(value=0, inplace=True)
df.total_rech_data_8.fillna(value=0, inplace=True)
df.total_rech_data_9.fillna(value=0, inplace=True)
df.av_rech_amt_data_6.fillna(value=0, inplace=True)
df.av_rech_amt_data_7.fillna(value=0, inplace=True)
df.av_rech_amt_data_8.fillna(value=0, inplace=True)
df.av_rech_amt_data_9.fillna(value=0, inplace=True)

```

```python
#Total recharge amounts for months 6 and 7
#Total recharge amount logic = Total data recharge + Total recharge Amount. 
#if any of the data recharge columns are 0 then retain the total recharge amt column as is

df['total_rech_amt_6'] = np.where((df['total_rech_data_6'] != 0) & (df['av_rech_amt_data_6'] != 0),
                                            df['total_rech_data_6']*df['av_rech_amt_data_6']+df['total_rech_amt_6'],
                                            df['total_rech_amt_6'])

df['total_rech_amt_7'] = np.where((df['total_rech_data_7'] != 0) & (df['av_rech_amt_data_7'] != 0),
                                            df['total_rech_data_7']*df['av_rech_amt_data_7']+df['total_rech_amt_7'],
                                            df['total_rech_amt_7'])
```

```python
# Filter high-value customers
df['av_rech_amt'] = (df["total_rech_amt_6"] + 
                          df["total_rech_amt_7"]) / 2.0
cutoff = df.av_rech_amt.quantile(.70)
print('70 percentile of first two months avg recharge amount: ', cutoff)
df1 = df[df['av_rech_amt'] >= cutoff]
```

```python
#Since we are not predicting any churn based on the above Date variables
#Derived Features based on the Exploratory Data Analysis performed in the previous step
#average ARPU during months 6 and 7

#average OG minutes of usage during months 6 and 7

#average IC Minutes of usage during months 6 and 7

#average OFFNET and ONNET minutes of usage for months 6 and 7

MLtelecom['gp_avg_arpu'] = (MLtelecom['arpu_6'] + MLtelecom['arpu_7'])/2
```

```python

#If a Customer is only joining in the Bad Phase assign the same valus to the Good Phase
MLtelecom['gp_avg_arpu'] = np.where((MLtelecom['arpu_8'] > 0) & (MLtelecom['gp_avg_arpu'] == 0),MLtelecom['arpu_8'],MLtelecom['gp_avg_arpu'])                              
```

```python
#Drop the individual month ARPU data as it's redundant
MLtelecom.drop(['arpu_6','arpu_7'],axis=1,inplace=True)
MLtelecom['total_og_mou_gp'] = (MLtelecom['total_og_mou_6'] + MLtelecom['total_og_mou_7'])/2
MLtelecom['total_og_mou_gp'] = np.where((MLtelecom['total_og_mou_8'] > 0) & (MLtelecom['total_og_mou_gp'] == 0),MLtelecom['total_ic_mou_8'],MLtelecom['total_og_mou_gp'])                              
```

```python
#Drop the individual month Outgoing usage data as it's redundant
MLtelecom.drop(['total_og_mou_6','total_og_mou_7'],axis=1,inplace=True)
MLtelecom['total_ic_mou_gp'] = (MLtelecom['total_ic_mou_6'] + MLtelecom['total_ic_mou_7'])/2

#If a Customer is only joining in the Bad Phase assign the same valus to the Good Phase
MLtelecom['total_ic_mou_gp'] = np.where((MLtelecom['total_ic_mou_8'] > 0) & (MLtelecom['total_ic_mou_gp'] == 0),MLtelecom['total_ic_mou_8'],MLtelecom['total_ic_mou_gp'])    
```

```python
#Drop the individual incoming usage month data as it's redundant
MLtelecom.drop(['total_ic_mou_6','total_ic_mou_7'],axis=1,inplace=True)

MLtelecom['onnet_mou_gp'] = (MLtelecom['onnet_mou_6'] + MLtelecom['onnet_mou_7'])/2
MLtelecom['offnet_mou_gp'] = (MLtelecom['offnet_mou_6'] + MLtelecom['offnet_mou_7'])/2
```

```python
#Drop the individual ONNET and OFFNET usage month data as it's redundant
MLtelecom.drop(['onnet_mou_6','onnet_mou_7','offnet_mou_6','offnet_mou_7'],axis=1,inplace=True)
```

```python
MLtelecom.fillna(0,inplace=True)
MLtelecom.shape
```

```python
#Feature Generation - Introduce new Feature called Retain Factor

#Retain Factor is calculated as the ratio of the Bad phase Average Revenue / Good Phase Average revenue
#And the Ratio of the number of recharges in month 8 Vs Month 7

#telecom_n.dtypes
MLtelecom['retain_factor_arpu'] = round(MLtelecom['arpu_8'] / MLtelecom['gp_avg_arpu'],2)
MLtelecom['retain_factor_rech'] = round(MLtelecom['total_rech_num_8'] / MLtelecom['total_rech_num_7'],2)
MLtelecom['retain_factor_rech'] = np.where(MLtelecom['retain_factor_rech'] > 1,1,MLtelecom['retain_factor_rech'])
MLtelecom['retain_factor_arpu'] = np.where(MLtelecom['retain_factor_arpu'] > 1,1,MLtelecom['retain_factor_arpu'])

```

```python
#Deduce a factor for retaining the customer
MLtelecom['retain_factor'] = np.where((MLtelecom['retain_factor_arpu'] > 0.5) & (MLtelecom['retain_factor_rech'] > 0.5),0,1)
MLtelecom.drop(columns = ['retain_factor_rech','retain_factor_arpu'], axis=1, inplace=True)
```
